```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gaussian
x = np.linspace(-4,4,10000)
y = 1/np.sqrt(2*np.pi) * np.exp(-.5*(x**2))
```

```python
fig, ax = plt.subplots(dpi = 500)
ax.plot(x,y)
plt.show()
```

```python
# uniform
x = np.linspace(-0.2,1.2,10000)
y = np.ones(x.shape[0])
y[x<0], y[x>1] = 0,0
fig, ax = plt.subplots(dpi = 500)
ax.plot(x,y, '--')
ax.scatter(x,y, s = 1)
plt.show()
```

```python
# geometric
x = np.arange(1,16)
p  = 0.35
y = (1-p)**(x-1)*p
fig, ax = plt.subplots(dpi = 500)
ax.stem(x,y)
plt.show()
```

```python
# cauchy
x = np.linspace(-5,5,10000)
y = 1/(np.pi*(1+x**2))
fig, ax = plt.subplots(dpi = 500)
ax.plot(x,y)
plt.show()
```
